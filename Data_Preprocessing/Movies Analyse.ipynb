 Cette fois-ci on cherche à savoir s'il y a un lien entre la catégorie de la vidéo et le mois d'apparition de la vidéo. Proposer une méthode et conclure.

```python
df_filtre['mois'] = pd.to_datetime(df_filtre['publish_time']).dt.month
df_filtre['mois'] = df_filtre['mois'].astype('str')

ct = pd.crosstab(df_filtre['category_id'], df_filtre['mois'])
```

```python
# TEST DE CHI-2

## Hypothèses : 

# 𝐻0 : La catégorie de la vidéo est indépendante du mois de l'apparition de la vidéo
# H1 : La catégorie de la vidéo  est dépendante du mois de l'apparition de la vidéo

from scipy.stats import chi2_contingency
resultats_chi2 = chi2_contingency(ct)

statistique = resultats_chi2[0]
p_valeur = resultats_chi2[1]
print("La statistique du test est : ", statistique, "\n"
      "La p-valeur du test est : ", p_valeur, "\n")

# Conclusions : 
print("Conclusion : p-val très petite (< 0.05) => on rejette H0 et on accepte H1.")
# on conclut qu'il y a un lien entre la catégorie de la vidéo et le mois d'apparition de la vidéo
```

```python
#  Afficher les indicateurs de position et de dispersion (lse plus usuels) de la variable "likes" du df.

# Indicateurs de positions

# Moyen
mean_likes = df['likes'].mean()

# Médiane
median_likes = df['likes'].median()

#quantiles
q1, q3 = df['likes'].quantile(q = [0.25, 0.75])

# Indicateurs de dispersion

# Ecart_types
std_likes = df['likes'].std()

# Interquartiles
IQR = q3 - q1

print ('Les indicateurs de position:', '\n'
      'Moyenne:',mean_likes, '\n'
      'Médiane:', median_likes, '\n'
      '1er quantile:', q1, '\n'
      '3eme quantile:',q3, '\n')
print('-----------------------------------------')

print('Les indicateurs de dispersion:', '\n'
     'Ecart_type:', std_likes, '\n'
     'Interquartiles:', IQR)
```

On voudrait savoir si le fait de désactiver les commentaires sur youtube a une influence sur les nombre de likes. Proposer une démarche scientifique, décrire les étapes et appliquer la méthode spécifique utilisé, puis conclure.



```python
df['comments_disabled'] = df['comments_disabled'].replace(('True','False'), (1,0))

df.head()
```

```python
## 1 - Les hypothèses :
# H0 : Le fait de désactiver les commentaires n'a pas d'influence sur le nombre de likes
# H1 : Le fait de désactiver les commentaires a une influence sur le nombre de likes

# 2 - le test statistique adapté, ici ANOVA :
import statsmodels.api 
result2 = statsmodels.formula.api.ols('likes ~ comments_disabled', data=df).fit()
table2 = statsmodels.api.stats.anova_lm(result2)
table2

# 3 - Conclusion : 
# p-val < 5% , on accepte H1
# on conclure que le fait de désactiver les commentaires a une influence sur le nombre de likes
```
