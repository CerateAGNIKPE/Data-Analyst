```python
df = pd.read_csv('CAvideos.csv')

df.head()
```

```python
# Afficher le type de chaque colonne et le changer s'il y a des variables qui ne sont pas dans un format adapt√©.


df.dtypes
```

```python
# Changer le format de la variable comments_disabled et publish_time 

df['likes'] = df['likes'].astype('int')

df['comments_disabled'] = df['comments_disabled'].astype('str')

# publish_time
df['publish_time'] = pd.to_datetime(df['publish_time'])

df.dtypes
```

```python

#Cr√©er un nouveau dataframe nomm√© df_filtre √† partir de df en gardant uniquement les cat√©gories Entertainment, Music, Gaming et Education. Afficher les  5

df_filtre = df[(df["category_id"] == 24) | (df["category_id"] == 10) | (df["category_id"] == 20) | (df["category_id"] == 27) ]
df_filtre.head()
```

```python
#Dans ce nouveau dataframe remplacer les nombres pr√©sents sur "categoryId" avec le nom de la cat√©gorie correspondante

df_filtre = df_filtre.replace((24,10,20,27), ('Entertainment', 'Music', 'Gaming', 'Education'))


df_filtre.head()
```

```python
df_filtre.dtypes
```

```python
groupby_m = df_filtre.groupby([pd.Grouper(key = 'publish_time', freq = 'm'), 
                               df_filtre['category_id']])\
                     .agg({'likes':'mean'}).unstack().fillna(0)

groupby_m.plot(figsize = (20, 4.5), style = 'o-');

# Les vid√©os de cat√©gories Entertaiment et Music ont plus de like que les vid√©os de cat√©gories Gaming et Education.
# Mais les vid√©os de cat√©gories Entertaiment ont plus de likes de fa√ßon g√©n√©rale
```

```python
df_filtre['mois'] = df_filtre['mois'].astype('int')
categories = df_filtre.groupby([df_filtre['mois'], 
                                df_filtre['category_id']])\
                      .agg({'video_id':'count'}).unstack().fillna(0)

categories.plot(figsize = (20, 4.5), style = 'o-');
```

 Cette fois-ci on cherche √† savoir s'il y a un lien entre la cat√©gorie de la vid√©o et le mois d'apparition de la vid√©o. Proposer une m√©thode et conclure.

```python
df_filtre['mois'] = pd.to_datetime(df_filtre['publish_time']).dt.month
df_filtre['mois'] = df_filtre['mois'].astype('str')

ct = pd.crosstab(df_filtre['category_id'], df_filtre['mois'])
```

```python
# TEST DE CHI-2

## Hypoth√®ses : 

# ùêª0 : La cat√©gorie de la vid√©o est ind√©pendante du mois de l'apparition de la vid√©o
# H1 : La cat√©gorie de la vid√©o  est d√©pendante du mois de l'apparition de la vid√©o

from scipy.stats import chi2_contingency
resultats_chi2 = chi2_contingency(ct)

statistique = resultats_chi2[0]
p_valeur = resultats_chi2[1]
print("La statistique du test est : ", statistique, "\n"
      "La p-valeur du test est : ", p_valeur, "\n")

# Conclusions : 
print("Conclusion : p-val tr√®s petite (< 0.05) => on rejette H0 et on accepte H1.")
# on conclut qu'il y a un lien entre la cat√©gorie de la vid√©o et le mois d'apparition de la vid√©o
```

```python
#  Afficher les indicateurs de position et de dispersion (lse plus usuels) de la variable "likes" du df.

# Indicateurs de positions

# Moyen
mean_likes = df['likes'].mean()

# M√©diane
median_likes = df['likes'].median()

#quantiles
q1, q3 = df['likes'].quantile(q = [0.25, 0.75])

# Indicateurs de dispersion

# Ecart_types
std_likes = df['likes'].std()

# Interquartiles
IQR = q3 - q1

print ('Les indicateurs de position:', '\n'
      'Moyenne:',mean_likes, '\n'
      'M√©diane:', median_likes, '\n'
      '1er quantile:', q1, '\n'
      '3eme quantile:',q3, '\n')
print('-----------------------------------------')

print('Les indicateurs de dispersion:', '\n'
     'Ecart_type:', std_likes, '\n'
     'Interquartiles:', IQR)
```

On voudrait savoir si le fait de d√©sactiver les commentaires sur youtube a une influence sur les nombre de likes. Proposer une d√©marche scientifique, d√©crire les √©tapes et appliquer la m√©thode sp√©cifique utilis√©, puis conclure.



```python
df['comments_disabled'] = df['comments_disabled'].replace(('True','False'), (1,0))

df.head()
```

```python
## 1 - Les hypoth√®ses :
# H0 : Le fait de d√©sactiver les commentaires n'a pas d'influence sur le nombre de likes
# H1 : Le fait de d√©sactiver les commentaires a une influence sur le nombre de likes

# 2 - le test statistique adapt√©, ici ANOVA :
import statsmodels.api 
result2 = statsmodels.formula.api.ols('likes ~ comments_disabled', data=df).fit()
table2 = statsmodels.api.stats.anova_lm(result2)
table2

# 3 - Conclusion : 
# p-val < 5% , on accepte H1
# on conclure que le fait de d√©sactiver les commentaires a une influence sur le nombre de likes
```
